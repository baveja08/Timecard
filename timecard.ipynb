```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("/Users/baveja/Desktop/timecard.csv")
df.head()
```

```python
df['Time']=pd.to_datetime(df['Time'])
df['Time Out']=pd.to_datetime(df['Time Out'])
df.head()
```

```python
df.sort_values(by=['Employee Name', 'Time'], inplace =True)
df.head()
```

```python
consecutive_days_7 = []
less_than_10_hours_between_shifts = []
more_than_14_hours_single_shift = []
```

```python
for employee, employee_data in df.groupby('Employee Name'):

        # Check for consecutive days worked
        consecutive_days = employee_data['Time'].diff().dt.days == 1
        consecutive_days_count = consecutive_days.astype(int).groupby(consecutive_days.cumsum()).cumsum()

        if any(consecutive_days_count >= 6):
            consecutive_days_7.append((employee, employee_data.iloc[0]['Position ID']))

        # Check for less than 10 hours between shifts
        time_between_shifts = (employee_data['Time'] - employee_data['Time Out'].shift()).dt.total_seconds() / 3600
        less_than_10_hours = (time_between_shifts < 10) & (time_between_shifts > 1)

        if any(less_than_10_hours):
            less_than_10_hours_between_shifts.append((employee, employee_data.iloc[0]['Position ID']))

        
```

```python
for employee, employee_data in df.groupby('Employee Name'):
    employee_data['Timecard Hours (as Time)'] = pd.to_numeric(employee_data['Timecard Hours (as Time)'], errors='coerce')

    
    # Check for more than 14 hours in a single shift
    more_than_14_hours = employee_data['Timecard Hours (as Time)'] > 14

    if any(more_than_14_hours):
        more_than_14_hours_single_shift.append((employee, employee_data.iloc[0]['Position ID']))

   
```

```python

print("Employees who have worked for 7 consecutive days:")
for employee, position_id in consecutive_days_7:
    print(f"{employee} - Position ID: {position_id}")


   
```

```python
print("\nEmployees who have less than 10 hours between shifts but greater than 1 hour:")
for employee, position_id in less_than_10_hours_between_shifts:
    print(f"{employee} - Position ID: {position_id}")
```

```python
print("\nEmployees who have worked for more than 14 hours in a single shift:")
for employee, position_id in more_than_14_hours_single_shift:
    print(f"{employee} - Position ID: {position_id}")
```
